```python
import pickle
from sklearn.metrics import r2_score, mean_absolute_error
from module.c_analysis import get_overall_metrics, show_pred_true

d_id = 'd00'
s_id = 's00'
m_id = 'm01'
e_id = 'e01'

o_results = get_overall_metrics(results, d_id, s_id, m_id, e_id)
handle = open(f'test_result/{d_id}_{s_id}_{m_id}_{e_id}.pickle', 'rb')
pred_tsty = pickle.load(handle)


for c_metrics, [pred, tsty] in zip(o_results.values, pred_tsty):
    case = c_metrics[0] 
    metrics = c_metrics[1:] 
    
    r2 = r2_score(tsty, pred)
    mae = mean_absolute_error(tsty, pred)
    
    p_title = f'{case} r2: {round(r2, 2)} mae: {round(mae, 2)}' # plot title
    
    show_pred_true(pred, tsty, 'sys', p_title, mode=1)
```
